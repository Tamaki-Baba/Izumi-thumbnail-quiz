### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
def interleave_strings(str1, str2):
    # 2つの文字列を交互に連結
    result = ''.join(a + b for a, b in zip(str1, str2))
    # 長さの異なる場合に備えて残りの文字を追加
    result += str1[len(str2):] + str2[len(str1):]
    return result

# 文字列の設定
str1 = "ヘルシンキ"
str2 = "エストニア"

# 結果の表示
interleaved_string = interleave_strings(str1, str2)
print(interleaved_string)
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
def word_length_list(sentence):
    # 文を単語に分解し、各単語の文字数をリストにする
    lengths = [len(word) for word in sentence.split()]
    return lengths

# 与えられた文
sentence = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"

# リストを作成
length_list = word_length_list(sentence)
print(length_list)
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
def create_word_dict(sentence, indices):
    # 文を単語に分解
    words = sentence.split()
    # 結果を格納する辞書
    result = {i: words[i-1][:2] for i in indices if i <= len(words)}
    return result

# 与えられた文
sentence = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."
# 取り出す単語の位置
indices = [1, 2, 5, 6, 8, 9, 16]

# 関数を呼び出して辞書を作成
word_dict = create_word_dict(sentence, indices)
print(word_dict)
```
